```python
import qrcode

# Data to encode
base = "tesouroQR-"

for i in range(11):
    if i ==0:
        add = "start"
    elif i < 10:
        add = f"0{i}"
    else:
        add = f"{i}"
    data = base + add    
    print(data)
    # Create QR code
    img = qrcode.make(data)
    img.save(data)

    

```
