```python
import qrcode

# Data to encode
base = "tesouroQR-"

for i in range(11):
    if i ==0:
        add = "start"
    elif i < 10:
        add = f"0{i}"
    else:
        add = f"{i}"
    data = base + add    
    print(data)
    # Create QR code
    img = qrcode.make(data)
    img.save(f"{data}.png")

    

```

```python
from reportlab.lib.pagesizes import A4
from reportlab.pdfgen import canvas
from reportlab.lib.units import inch
from PIL import Image
import glob, os, io

# Collect and sort PNGs
images = sorted(glob.glob("*.png"))

# Prepare PDF
c = canvas.Canvas("QR_sheet.pdf", pagesize=A4)
page_w, page_h = A4

cols, rows = 2, 5
qr_size = 2 * inch
x_margin = 1 * inch
y_margin = 0.8 * inch
x_spacing = (page_w - 2 * x_margin - cols * qr_size) / (cols - 1)
y_spacing = (page_h - 2 * y_margin - rows * qr_size) / (rows - 1)

for i, img_path in enumerate(images[:10]):
    # Open with Pillow and ensure RGB (no alpha/transparency)
    with Image.open(img_path) as im:
        im = im.convert("RGB")
        buf = io.BytesIO()
        im.save(buf, format="PNG")
        buf.seek(0)

        col = i % cols
        row = i // cols
        x = x_margin + col * (qr_size + x_spacing)
        y = page_h - y_margin - (row + 1) * (qr_size + y_spacing)

        c.drawImage(buf, x, y, qr_size, qr_size, preserveAspectRatio=True, mask='auto')
        label = os.path.splitext(os.path.basename(img_path))[0]
        c.setFont("Helvetica", 9)
        c.drawCentredString(x + qr_size / 2, y - 12, label)

c.save()
print("✅ PDF created successfully: QR_sheet.pdf")
```
