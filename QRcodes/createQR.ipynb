```python
import qrcode

# Data to encode
base = "tesouroQR-"

for i in range(11):
    if i ==0:
        add = "start"
    elif i < 10:
        add = f"0{i}"
    else:
        add = f"{i}"
    data = base + add    
    print(data)
    # Create QR code
    img = qrcode.make(data)
    img.save(f"{data}.png")
```

```python
from reportlab.lib.pagesizes import A4
from reportlab.pdfgen import canvas
from reportlab.lib.units import inch
from reportlab.lib.utils import ImageReader
from PIL import Image
import glob, os, io

# Collect and sort images
images = sorted(glob.glob("*.png"))

# Setup PDF
page_w, page_h = A4
c = canvas.Canvas("QR_sheet.pdf", pagesize=A4)

# Layout constants
cols, rows = 2, 5
qr_size = 2 * inch
x_margin = 1 * inch
y_top = page_h - 1.25 * inch
x_gap = 3.5 * inch     # horizontal spacing
y_gap = 1.9 * inch     # vertical spacing between rows

# Precompute coordinates
positions = []
for i, img_path in enumerate(images[:10]):
    row = i // cols
    col = i % cols
    x = x_margin + col * x_gap
    y = y_top - row * y_gap - qr_size
    positions.append((img_path, x, y))

# Pass 1 — draw all QR codes
for img_path, x, y in positions:
    with Image.open(img_path) as im:
        im = im.convert("RGB")
        buf = io.BytesIO()
        im.save(buf, format="PNG")
        buf.seek(0)
        img_reader = ImageReader(buf)
        c.drawImage(img_reader, x, y, qr_size, qr_size, preserveAspectRatio=True, mask='auto')

# Pass 2 — draw all labels below QRs
for img_path, x, y in positions:
    label = os.path.splitext(os.path.basename(img_path))[0]
    c.setFont("Helvetica", 10)
    c.drawCentredString(x + qr_size / 2, y + 10 , label)

# Save the final PDF
c.save()
print("✅ Created QR_sheet.pdf successfully — all 10 codes and labels visible!")
```
